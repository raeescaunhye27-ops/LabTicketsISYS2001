## Part 3: Smart Purchase Advisor (20 minutes)

### The Challenge
Build a program that helps users make smart purchase decisions by considering multiple factors.

### Requirements

Your purchase advisor should:

1. Get purchase amount and user's available money
2. Check if they can afford it (basic affordability)
3. Give different advice based on purchase size
4. Consider their financial situation

### Starter Framework

````python
print("=== Smart Purchase Advisor ===")

# Get financial information
available_money = float(input("How much money do you have available? $"))
purchase_price = float(input("How much does the item cost? $"))
monthly_income = float(input("What's your monthly income? $"))

# Calculate affordability percentage
affordability_ratio = purchase_price / available_money

# Decision logic
if purchase_price > available_money:
    print("You cannot afford this purchase right now.")
    print("Advice: Save more money before buying.")

elif affordability_ratio > 0.7:
    print("Warning: This purchase will use most of your available money.")
    print("Advice: Think carefully before buying.")

elif purchase_price > 0.5 * monthly_income:
    print("This is quite expensive compared to your monthly income.")
    print("Advice: Consider waiting or budgeting for it.")

elif affordability_ratio > 0.3:
    print("This purchase is reasonable but still significant.")
    print("Advice: Make sure it fits your financial priorities.")

else:
    print("Good news! This purchase seems affordable.")
    print("Advice: You can buy it without major financial stress.")
````

````python
print("=== Personal Finance Tracker v0.1 ===")
print("Week 3: Smart Budget Decisions")
print()

# User setup
name = input("Enter your name: ")
monthly_income = float(input("Enter your monthly income: $"))

print(f"\nWelcome to your finance tracker, {name}!")
print(f"Monthly income: ${monthly_income:.2f}")

# This week's feature: Expense categorisation
print("\n--- Expense Entry ---")
expense_amount = float(input("Enter an expense: $"))
expense_description = input("What was this expense for? ")

# Calculate how big the expense is compared to income
expense_ratio = expense_amount / monthly_income

# Decision making
if expense_ratio >= 0.30:
    category = "Major Expense"
    advice = "This is a large expense. Make sure it is necessary and planned."

elif expense_ratio >= 0.15:
    category = "Significant Expense"
    advice = "This is a noticeable expense. Keep track of your budget."

elif expense_ratio >= 0.05:
    category = "Moderate Expense"
    advice = "This expense is manageable but should still be monitored."

else:
    category = "Minor Expense"
    advice = "Small purchase. It fits easily within your budget."

# Display result
print("\n--- Expense Summary ---")
print(f"Expense: {expense_description}")
print(f"Amount: ${expense_amount:.2f}")
print(f"Category: {category}")
print(f"Advice: {advice}")
````
