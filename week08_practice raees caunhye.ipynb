## Practice Exercise 1.2: Top Performing Products

::: {.callout-tip}
## Business Scenario

Your inventory manager needs to know which products to stock more of next quarter.

**Challenge**: Ask AI to identify the top 5 products by:
- Total revenue generated
- Total units sold
- Which products have the highest profit margins (hint: compare unit price trends)

**Practice better prompting**: Be specific about formatting and business context.
:::

**Your Prompt:**
```
["Analyze the sales and inventory dataset to identify the top 5 performing products for next quarter stocking decisions. For each product, calculate:]
```

**AI Response:**

````python
import pandas as pd

# Load dataset
# Replace 'sales_data.csv' with your file name
df = pd.read_csv('sales_data.csv')

# Preview data
print(df.head())

# --- Ensure required columns exist ---
# Expected columns: Product, Units_Sold, Unit_Price, Cost_Price (if available)

# Create Revenue column
df['Revenue'] = df['Units_Sold'] * df['Unit_Price']

# Create Profit (if cost price exists)
if 'Cost_Price' in df.columns:
    df['Profit'] = (df['Unit_Price'] - df['Cost_Price']) * df['Units_Sold']
    df['Profit_Margin'] = (df['Unit_Price'] - df['Cost_Price']) / df['Unit_Price']
else:
    # Approximate margin using price trend (fallback)
    df['Profit_Margin'] = df['Unit_Price'] / df['Unit_Price'].max()

# --- Aggregate by Product ---
summary = df.groupby('Product').agg(
    Total_Revenue=('Revenue', 'sum'),
    Total_Units_Sold=('Units_Sold', 'sum'),
    Avg_Profit_Margin=('Profit_Margin', 'mean')
).reset_index()

# --- Rank Products ---
top_products = summary.sort_values(
    by=['Total_Revenue', 'Total_Units_Sold'],
    ascending=False
).head(5)

# --- Display Results ---
print("\nTop 5 Performing Products:\n")
print(top_products)

# --- Optional: Insights ---
print("\nInsights:")
for index, row in top_products.iterrows():
    print(f"- {row['Product']} performs strongly with revenue {row['Total_Revenue']} "
          f"and {row['Total_Units_Sold']} units sold.")
````

**Business Insight Questions:**
- Which products should the company stock more of?
The company should stock more of the top-performing products that generate high total revenue and have strong unit sales. Products that combine high demand (volume) with good profit margins are the most valuable and should be prioritized for restocking.
- Are high-price items selling well, or is volume more important?
In most cases, volume is more important than high price. Products with lower or moderate prices often sell in higher quantities, contributing more to total revenue. However, some high-price items may still perform well if they maintain strong margins, so a balance is important.
- What would you recommend to the inventory manager?
I would recommend:

Prioritizing high-demand, high-revenue products for restocking
Maintaining a mix of high-volume items and high-margin premium products
Monitoring pricing trends and customer demand regularly
Reducing stock of low-performing or slow-moving products
Using data insights to plan future inventory and promotions
